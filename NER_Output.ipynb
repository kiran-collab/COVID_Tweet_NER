```python
from flair.data import Sentence
from flair.models import SequenceTagger
sentence=Sentence('Sam please send team india immediately america sanitize building')
tagger=SequenceTagger.load('ner')
tagger.predict(sentence)
print(sentence)
print('The following NER tags are found:')
for entity in sentence.get_spans('ner'):
  print(entity)
```

```python
import time
import csv
from flair.data import Sentence
from flair.models import SequenceTagger
from csv import writer
from csv import reader
count=1
# load the NER tagger
tagger = SequenceTagger.load('ner')

with open(r'/content/drive/MyDrive/COVID_Datasets/tweet.txt.txt') as tfile, open('/content/drive/MyDrive/COVID_Datasets/COVID_tweets_entities.csv', 'w') as ofile:
    while (True):
      text=tfile.readline()
      if not text:
        break
      # make a sentence
      sentence = Sentence(text)
      # run NER over sentence
      tagger.predict(sentence)
      print(sentence)
      print('The following NER tags are found:')
      for entity in sentence.get_spans('ner'):
        print(entity)
        csv_writer = writer(ofile)
        # Write the updated row / list to the output file
        csv_writer.writerow(entity)
```
